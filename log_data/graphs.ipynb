```python
import numpy as np
from scipy.signal import savgol_filter 
import pandas as pd
import matplotlib.pyplot as plt

import os
```

```python
def smooth(y, window, poly=1):
    '''
    y: vector to be smoothed 
    window: size of the smoothing window '''
    return savgol_filter(y,window,poly)

def get_lengths(path,player):
    data = pd.read_csv(path)
    y = data['Value'].to_numpy()
    y = y*2
    if(player =="O"):
        y = y+1
    y = smooth(y,50)

    return y, np.array([1024*i for i in range(1,len(y)+1)])

def get_rewards(path):
    data = pd.read_csv(path)
    y = data['Value'].to_numpy()
    y = smooth(y,50)

    return y, np.array([1024*i for i in range(1,len(y)+1)])
```

```python

plt.figure(figsize=(6, 4))


path="log_data\mean_length_V3"

for file in os.listdir(path):
    file_path = os.path.join(path,file)
    player = "O" if "O" in file else "X"
    env = "TT" if "TT" in file else "TF" if "TF" in file else "FT"
    x,y = get_lengths(file_path,player)
    
    plt.plot(y,x,label=f"{player}_{env}")

plt.ylim((0,32))
plt.xlabel("Timestemp")
plt.ylabel("Average game length")
# plt.title("Quantum Rules V3")
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1),
          fancybox=True, shadow=True, ncol=3)
plt.plot()
```

```python
plt.figure(figsize=(6, 4))


path="log_data\mean_reward_V3"

for file in os.listdir(path):
    file_path = os.path.join(path,file)
    player = "O" if "O" in file else "X"
    env = "TT" if "TT" in file else "TF" if "TF" in file else "FT"
    x,y = get_rewards(file_path)
    
    plt.plot(y,x,label=f"{player}_{env}")

plt.ylim((-1,1))
plt.xlabel("Timestemp")
plt.ylabel("Average reward")
# plt.title("Quantum Rules V3")
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1),
          fancybox=True, shadow=True, ncol=3)
plt.plot()
```
