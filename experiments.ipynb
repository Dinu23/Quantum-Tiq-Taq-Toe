```python
from play import play_agents
from agents import RandomAgent,ModelAgent
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
#constant LR

matrix = np.zeros((15,15))
N = 100
path = "models10k1000\model_10000_"

for i in range(15):
    for j in range(15):
        if(i == 0):
            player_X = RandomAgent()
        else:
            player_X = ModelAgent(f"{path}{min(i*15,200)}")
        if(j == 0):
            player_O = RandomAgent()
        else:
            player_O = ModelAgent(f"{path}{min(j*15,200)}")

        for _ in range(N):
            matrix[i][j] += play_agents(player_X,player_O, render_mode=None)
        matrix[i][j] /=N
        print(i,j,"->", matrix[i][j])

np.save("results_constant_lr",matrix)
```

```python
from scipy.signal import savgol_filter 
import pandas as pd

def smooth(y, window, poly=1):
    '''
    y: vector to be smoothed 
    window: size of the smoothing window '''
    return savgol_filter(y,window,poly)

def get_list(path,r=1):
    data = pd.read_csv(path)
    y = data['Value'].to_numpy()
    
    y = smooth(y,50)

    return y, np.array([r*i for i in range(1,len(y)+1)])


```

```python

plt.figure(figsize=(14, 6))

x,y = get_list("data\\run-O-2a-tag-rollout_ep_rew_mean.csv")
plt.plot(y*1024,x,label="O-2a")
x,y = get_list("data\\run-X-2a-tag-rollout_ep_rew_mean.csv")
plt.plot(y*1024,x,label="X-2a")


x,y = get_list("data\\run-O-lrc-tag-rollout_ep_rew_mean.csv",r=2)
plt.plot(y*1048,x,label="O-lrc")
x,y = get_list("data\\run-X-lrc-tag-rollout_ep_rew_mean.csv",r=2)
plt.plot(y*1048,x,label="X-lrc")


x,y = get_list("data\\run-O-lrl-tag-rollout_ep_rew_mean.csv")
plt.plot(y*1024,x,label="O-lrl")
x,y = get_list("data\\run-X-lrl-tag-rollout_ep_rew_mean.csv")
plt.plot(y*1024,x,label="X-lrl")

x,y = get_list("data\\run-O-net-tag-rollout_ep_rew_mean.csv")
plt.plot(y*1024,x,label="O-net")
x,y = get_list("data\\run-X-net-tag-rollout_ep_rew_mean.csv")
plt.plot(y*1024,x,label="X-net")

plt.xlabel("Timestemp")
plt.ylabel("Average return")

plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1),
          fancybox=True, shadow=True, ncol=4)
plt.title("Average return during training")
```
